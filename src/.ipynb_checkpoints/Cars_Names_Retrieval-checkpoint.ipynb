# Cars Manufacturers and Models Rerieval

retrieval from https://www.auto-data.net/it/

```python
from bs4 import BeautifulSoup
import requests
```

```python
seed = 'https://www.auto-data.net'
page = requests.get(seed)
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
cars = {}
```

```python
manufacturers = []
for i in range(len(soup.find_all('a', 'marki_blok'))):
    manuf = soup.find_all('a', 'marki_blok')[i].h2.text
    url = seed + soup.find_all('a', 'marki_blok')[i].get('href')
    models_page = requests.get(url)
    models_soup = BeautifulSoup(models_page.text, 'html.parser')
    models = []
    for j in range(len(models_soup.find_all('a', 'modeli'))):
        models.append(models_soup.find_all('a', 'modeli')[j].h2.text)
    cars[manuf] = models
```

```python
cars
```

Save into file

```python
f = open("resources/cars.txt", "a")
for man in cars:
    for c in cars[man]:
        f.write('(' + man + ',' + c + ')\n')
f.close()
```
