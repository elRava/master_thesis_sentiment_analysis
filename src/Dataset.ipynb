### All imports

```python
import glob
import matplotlib
import pandas as pd
import numpy as np
```

## Import data from files

```python
#combine all files in the list
all_filenames =  sorted(glob.glob('labeling/fatti/finiti/csv/*.csv'))
dataset = pd.concat([pd.read_csv(f, sep=';', encoding = "ISO-8859-1", low_memory=False) for f in all_filenames ], ignore_index=True)
```

```python
# remove all NaN
dataset['(Testo Citato)'] = dataset['(Testo Citato)'].replace(np.nan, "")
dataset['TESTO'] = dataset['TESTO'].replace(np.nan, "")
dataset['Costruttore'] = dataset['Costruttore'].replace(np.nan, "")
dataset['Modello'] = dataset['Modello'].replace(np.nan, "")
# all left cells are considered neutral
# Meccanica;Prestazioni;Consumi/Autonomia;Motore;Elettrico;Fuoristrada;Elettronica di Bordo;Manutenzione;Assistenza/Concessionari
dataset['Sentiment'] = dataset['Sentiment'].replace(np.nan, "neutro")
dataset['Brand'] = dataset['Brand'].replace(np.nan, "irrilevante")
dataset['Interni'] = dataset['Interni'].replace(np.nan, "irrilevante")
dataset['Esterni'] = dataset['Esterni'].replace(np.nan, "irrilevante")
dataset['Meccanica'] = dataset['Meccanica'].replace(np.nan, "irrilevante")
dataset['Prestazioni'] = dataset['Prestazioni'].replace(np.nan, "irrilevante")
dataset['Consumi/Autonomia'] = dataset['Consumi/Autonomia'].replace(np.nan, "irrilevante")
dataset['Motore'] = dataset['Motore'].replace(np.nan, "irrilevante")
dataset['Elettrico'] = dataset['Elettrico'].replace(np.nan, "irrilevante")
dataset['Fuoristrada'] = dataset['Fuoristrada'].replace(np.nan, "irrilevante")
dataset['Elettronica di Bordo'] = dataset['Elettronica di Bordo'].replace(np.nan, "irrilevante")
dataset['Manutenzione'] = dataset['Manutenzione'].replace(np.nan, "irrilevante")
dataset['Assistenza/Concessionari'] = dataset['Assistenza/Concessionari'].replace(np.nan, "irrilevante")
```

```python
# show dataset
dataset
```

Clean Dataset

```python
# all lowercase
for i in range(dataset.shape[0]):
    dataset.loc[i, 'Costruttore'] = dataset.loc[i, 'Costruttore'].lower()
    dataset.loc[i, 'Modello'] = dataset.loc[i, 'Modello'].lower()
```

```python
# remove last character spaces
for i in range(dataset.shape[0]):
    if dataset.loc[i, 'Costruttore'] != '' and dataset.loc[i, 'Costruttore'][-1] == ' ':
        dataset.loc[i, 'Costruttore'] = dataset.loc[i, 'Costruttore'][:-1]
    if dataset.loc[i, 'Modello'] != '' and dataset.loc[i, 'Modello'][-1] == ' ':
        dataset.loc[i, 'Modello'] = dataset.loc[i, 'Modello'][:-1]
```

```python
# correct mispellings manually
for i in range(dataset.shape[0]):
    if dataset.loc[i, 'Costruttore'] in ['alfa', 'alfaromeo']:
        dataset.loc[i, 'Costruttore'] = 'alfa romeo'
    elif dataset.loc[i, 'Costruttore'] in ['ausi']:
        dataset.loc[i, 'Costruttore'] = 'audi'
    elif dataset.loc[i, 'Costruttore'] in ['bmv']:
        dataset.loc[i, 'Costruttore'] = 'bmw'
    elif dataset.loc[i, 'Costruttore'] in ['chervolet', 'chrvrolet']:
        dataset.loc[i, 'Costruttore'] = 'chevrolet'
    elif dataset.loc[i, 'Costruttore'] in ['citroën']:
        dataset.loc[i, 'Costruttore'] = 'citroen'
    elif dataset.loc[i, 'Costruttore'] in ['hyundau']:
        dataset.loc[i, 'Costruttore'] = 'hyundai'
    elif dataset.loc[i, 'Costruttore'] in ['infinity', 'infility']:
        dataset.loc[i, 'Costruttore'] = 'infiniti'
    elif dataset.loc[i, 'Costruttore'] in ['lamborighini']:
        dataset.loc[i, 'Costruttore'] = 'lamborghini'
    elif dataset.loc[i, 'Costruttore'] in ['mc laren']:
        dataset.loc[i, 'Costruttore'] = 'mclaren'
    elif dataset.loc[i, 'Costruttore'] in ['mecedes-benz', 'mercedes benz', 'mercedes-benz', 'mercedes_benz']:
        dataset.loc[i, 'Costruttore'] = 'mercedes'
    elif dataset.loc[i, 'Costruttore'] in ['mitsubushi', 'mitshubishi']:
        dataset.loc[i, 'Costruttore'] = 'mitsubishi'
    elif dataset.loc[i, 'Costruttore'] in ['oper']:
        dataset.loc[i, 'Costruttore'] = 'opel'
    elif dataset.loc[i, 'Costruttore'] in ['pegeuot', 'peugeaut', 'peugeout', 'ppeugeot']:
        dataset.loc[i, 'Costruttore'] = 'peugeot'
    elif dataset.loc[i, 'Costruttore'] in ['porsceh', 'posche']:
        dataset.loc[i, 'Costruttore'] = 'porsche'
    elif dataset.loc[i, 'Costruttore'] in ['subari', 'subau', 'subreza']:
        dataset.loc[i, 'Costruttore'] = 'subaru'
    elif dataset.loc[i, 'Costruttore'] in ['suzuky']:
        dataset.loc[i, 'Costruttore'] = 'suzuki'
    elif dataset.loc[i, 'Costruttore'] in ['2. vw', 'volkvagen', 'volkwagen', 'volsfwagen', 'volswagen', 'volwagen', 'vw', 'wolkvagen']:
        dataset.loc[i, 'Costruttore'] = 'volkswagen'
    elif dataset.loc[i, 'Costruttore'] in ['laxus']:
        dataset.loc[i, 'Costruttore'] = 'lexus'
    elif dataset.loc[i, 'Costruttore'] in ['daiatsu']:
        dataset.loc[i, 'Costruttore'] = 'daihatsu'
    elif dataset.loc[i, 'Costruttore'] in ['reanult', 'renaul', 'renaule', 'renult']:
        dataset.loc[i, 'Costruttore'] = 'renault'
    elif dataset.loc[i, 'Costruttore'] in ['Range Rover']:
        dataset.loc[i, 'Costruttore'] = 'Land Rover'
    elif dataset.loc[i, 'Costruttore'] in [np.nan]:
        dataset.loc[i, 'Costruttore'] = ''
    # confusion on brand and model
    elif dataset.loc[i, 'Costruttore'] in ['avensis']:
        dataset.loc[i, 'Costruttore'] = 'toyota'
        dataset.loc[i, 'Modello'] = 'avensis'
    elif dataset.loc[i, 'Costruttore'] in ['golf']:
        dataset.loc[i, 'Costruttore'] = 'volkswagen'
        dataset.loc[i, 'Modello'] = 'golf'
    elif dataset.loc[i, 'Costruttore'] in ['passat']:
        dataset.loc[i, 'Costruttore'] = 'volkswagen'
        dataset.loc[i, 'Modello'] = 'passat'
    elif dataset.loc[i, 'Costruttore'] in ['corvette']:
        dataset.loc[i, 'Costruttore'] = 'chevrolet'
        dataset.loc[i, 'Modello'] = 'corvette'
    elif dataset.loc[i, 'Costruttore'] in ['lancer']:
        dataset.loc[i, 'Costruttore'] = 'mitsubishi'
        dataset.loc[i, 'Modello'] = 'lancer'
    elif dataset.loc[i, 'Costruttore'] in ['leon']:
        dataset.loc[i, 'Costruttore'] = 'seat'
        dataset.loc[i, 'Modello'] = 'leon'
    elif dataset.loc[i, 'Costruttore'] in ['logan']:
        dataset.loc[i, 'Costruttore'] = 'dacia'
        dataset.loc[i, 'Modello'] = 'logan'
    elif dataset.loc[i, 'Costruttore'] in ['mazda 3']:
        dataset.loc[i, 'Costruttore'] = 'mazda'
        dataset.loc[i, 'Modello'] = '3'
    elif dataset.loc[i, 'Costruttore'] in ['nissan gtr']:
        dataset.loc[i, 'Costruttore'] = 'nissan'
        dataset.loc[i, 'Modello'] = 'gt-r'
    elif dataset.loc[i, 'Costruttore'] in ['scenic']:
        dataset.loc[i, 'Costruttore'] = 'renault'
        dataset.loc[i, 'Modello'] = 'scenic'
    elif dataset.loc[i, 'Costruttore'] in ['yaris']:
        dataset.loc[i, 'Costruttore'] = 'toyota'
        dataset.loc[i, 'Modello'] = 'yaris'
    elif dataset.loc[i, 'Costruttore'] in ['punto']:
        dataset.loc[i, 'Costruttore'] = 'fiat'
        dataset.loc[i, 'Modello'] = 'punto'
    elif dataset.loc[i, 'Costruttore'] in ['audi a4']:
        dataset.loc[i, 'Costruttore'] = 'audi'
        dataset.loc[i, 'Modello'] = 'a4'
    elif dataset.loc[i, 'Costruttore'] in ['fx']:
        dataset.loc[i, 'Costruttore'] = 'infiniti'
        dataset.loc[i, 'Modello'] = 'fx'
```

## Statistics

For each category count the number of positive/negatiive/neutral/irrelevant comments

```python
cols = ['Sentiment', 'Brand', 'Interni', 'Esterni', 'Meccanica', 'Prestazioni', 'Consumi/Autonomia', 'Motore', 'Elettrico', 'Fuoristrada', 'Elettronica di Bordo', 'Manutenzione', 'Assistenza/Concessionari']
```

### Complete count

```python
stats_complete = []
for c in cols:
    stats_complete.append([list(dataset[c]).count('molto positivo'), list(dataset[c]).count('positivo'), list(dataset[c]).count('neutro'), list(dataset[c]).count('negativo'), list(dataset[c]).count('molto negativo'), list(dataset[c]).count('irrilevante')])
# pandas DataFrame of stats
cols_complete = ['very positive', 'positive', 'neutral', 'negative' , 'very negative', 'irrelevant']
statistics_complete = pd.DataFrame(stats_complete, columns=cols_complete, index=cols) 
statistics_complete
```

### Compact count

```python
stats_compact = []
for c in cols:
    stats_compact.append([list(dataset[c]).count('molto positivo') + list(dataset[c]).count('positivo'), list(dataset[c]).count('neutro'), list(dataset[c]).count('negativo') + list(dataset[c]).count('molto negativo'), list(dataset[c]).count('irrilevante')])
# pandas DataFrame of stats
cols_compact = ['positive', 'neutral', 'negative', 'irrelevant']
statistics_compact = pd.DataFrame(stats_compact, columns=cols_compact, index=cols) 
statistics_compact
```
